Task 1:- Prediction using Decision Tree algorithm

```python
from matplotlib import pyplot as plt
from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree

data = datasets.load_iris()
X = data.data
Y = data.target

print(data)
```

```python
classifying_tree = DecisionTreeClassifier(random_state=0)
model = classifying_tree.fit(X,Y)
my_tree = tree.export_text(classifying_tree)
print(my_tree)
```

```python

fig = plt.figure(figsize = (15,15))

vis_tree = tree.plot_tree(classifying_tree, feature_names = data.feature_names, class_names = data.target_names, filled = True)
```
